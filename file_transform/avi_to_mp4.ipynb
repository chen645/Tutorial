```python
import cv2
def convert_avi_to_mp4(avi,mp4):
    videoCapture = cv2.VideoCapture(avi)  

    #获得码率及尺寸  
    fps = videoCapture.get(cv2.CAP_PROP_FPS)  
    size = (int(videoCapture.get(cv2.CAP_PROP_FRAME_WIDTH)),   
            int(videoCapture.get(cv2.CAP_PROP_FRAME_HEIGHT)))  

    #指定写视频的格式, I420-avi, MJPG-mp4  
    videoWriter = cv2.VideoWriter(mp4, cv2.VideoWriter_fourcc('M', 'J', 'P', 'G'), fps, size)  

    #读帧  
    success, frame = videoCapture.read()  

    while success :  
        
        videoWriter.write(frame) #写视频帧  
        success, frame = videoCapture.read() #获取下一帧  
```

```python
import os

data_path = "subject_1"
dest_path="emotion_mp4"
for emotion in sorted(os.listdir(data_path)):
    for mp4 in sorted(os.listdir(data_path+'/'+emotion)):
        print(mp4)
        convert_avi_to_mp4(data_path+'/'+emotion+'/'+mp4,dest_path+'/'+emotion+'/'+mp4[:-4]+'.mp4')
        
```
